```python
%matplotlib inline
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
plt.rcParams["figure.figsize"] = (12,8)
```

**0.**

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

**1a.**

```python
fig,ax = plt.subplots(1,1)
plt.plot(t[:1000],bkgd[:1000],'o')
plt.show()
```

**1b.**

```python
def plotXPoints(X,x0):
    fig,ax = plt.subplots(1,1)
    p_t = t[:(x0+X)][x0:]
    p_bkgd = bkgd[:(x0+X)][x0:]
    plt.plot(p_t,p_bkgd,'o')
    plt.title("Scatter Plot of " + str(X) + " Points starting from Point #" + str(x0),fontsize=16)
    plt.show()

plotXPoints(2000,50000)
```

**1c.**

```python
p_t = np.zeros(1000)
p_bkgd = np.zeros(1000)
for i in range(0,1000):
    p_t[i] = t[i*100]
    p_bkgd[i] = bkgd[i*100]

fig,ax = plt.subplots(1,1)
plt.plot(p_t,p_bkgd,'o')
plt.title("Scatter Plot of every 100th point",fontsize=16)
plt.show()
```

**2a.**

```python
fig,ax = plt.subplots(1,1)
plt.hist2d(t,bkgd,120)
plt.colorbar()
plt.title("2D Histogram of All Points",fontsize=16)
plt.show()
```

**2b.** The color of each bin in this plot represents the number of points inside of that bin, that number can be seen from the color bar. This is plotted over 120x120 bins, where the x-axis is the index of measurement (ex. time) and the y-axis is the measurement itself.<br>

**3a.**

```python
p_t = np.zeros(len(t))
for i in range(0,len(t)):
    p_t[i] = t[i]%(2*math.pi)

fig,ax = plt.subplots(1,1)
plt.plot(p_t,bkgd,'o')
plt.title("Scatter Plot of All Points folded over 2$\pi$",fontsize=16)
plt.show()
```

**3b.**

```python
fig,ax = plt.subplots(1,1)
plt.hist2d(p_t,bkgd,120)
plt.colorbar()
plt.title("2D Histogram of All Points folded by 2$\pi$",fontsize=16)
plt.show()
```

Note how there's a clear one-period sinusoidal signal now.
