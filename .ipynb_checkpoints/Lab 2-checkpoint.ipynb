```python
%matplotlib inline
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats, signal


plt.rcParams["figure.figsize"] = (15,10)
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(0, 10, 11)
plt.step(x, stats.poisson.pmf(x, 5), where='post')
plt.show()
```

```python
def conv_poi(n):
    x = np.linspace(0, 10, 11)
    base = stats.poisson.pmf(x, 5)
    for i in range(n-1):
        base = signal.convolve(base, stats.poisson.pmf(x, 5))
    x = np.linspace(0, 10*n, (10*n)+1)
    return x, base
```

```python
fig, ax = plt.subplots(1, 1)
x, pmf = conv_poi(5)
plt.step(x, pmf, where='post', linewidth='7')
plt.step(x, stats.poisson.pmf(x, 25), where='post', color='orange')
plt.show()
```

```python
days = [2, 5, 10, 20]
fig, ax = plt.subplots(4,figsize=(15,30))
for i in range(len(days)):
    x, pmf = conv_poi(days[i])
    ax[i].step(x, pmf, where='post', linewidth='7')
    ax[i].step(x, stats.poisson.pmf(x, 5*days[i]),where='post', color='orange')
plt.show()
```

```python
def conv_ray(n):
    x = np.linspace(0, 10, 11)
    base = stats.rayleigh.pdf(x, 0, 5)
    for i in range(n-1):
        base = np.convolve(base, stats.rayleigh.pdf(x, 0, 5))
    x = np.linspace(0, 10*n, 1000)
    return x, base
```
